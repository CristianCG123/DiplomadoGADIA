Bucle for El bucle for se utiliza para iterar sobre una secuencia (como una lista, tupla, diccionario, conjunto o cadena de caracteres). Se ejecuta una vez por cada elemento en la secuencia.

```python
# imprimir todos los números de una secuencia de datos
secuencia =[1,2,3,4,5,]

for mumero in secuencia:

  print(mumero)
```

```python
for numero in secuencia:
  if numero % 2 == 0:
    print(f"el número {numero} es par")
  else:
    print(f"el número {numero} es inpar")
```

Supongamos que quieres solicitar a un usuario que ingrese una contraseña hasta que lo haga correctamente.


```python
contraseña_correct="123"
intento = 3

while intento > 0:
    contraseña = input("introduce tu contraseña: ")
    if contraseña == contraseña_correct:
      print("Aceeso concedido")
      break
    else :
      intento -=1
      print(f"Contraseña incorrecta, te quedan {intento} intentos")
if intento == 0:

  print("Ya no te quedan intentos")

```

```python
import random

# Definir el número secreto
numero_secreto = random.randint(1, 10)

print("¡Adivina el número secreto entre 1 y 10!")

# Variable para almacenar la adivinanza del usuario
adivinanza = None

# Bucle para seguir pidiendo al usuario que adivine
while adivinanza != numero_secreto:
    adivinanza = int(input("Introduce tu número: "))
    if adivinanza < numero_secreto:
        print("¡Demasiado bajo! Intenta de nuevo.")
    elif adivinanza > numero_secreto:
        print("¡Demasiado alto! Intenta de nuevo.")

# Mensaje de felicitación cuando acierta
print(f"¡Felicidades! Has adivinado el número secreto: {numero_secreto}")
```
